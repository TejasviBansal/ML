## Identified Questions and Answers

### Question 1: What is unsupervised learning in the context of machine learning?
**Answer:** Unsupervised learning is a type of machine learning that deals with unlabeled data. The goal is to discover hidden patterns, structures, or relationships within the data without any prior knowledge of output variables. Common tasks include clustering, dimensionality reduction, and anomaly detection.

### Question 2: How does K-Means clustering algorithm work?
**Answer:** K-Means clustering works by iteratively partitioning `n` data points into `k` clusters. It starts by randomly initializing `k` centroids. Then, it assigns each data point to the nearest centroid, forming `k` clusters. After assignment, the centroids are re-calculated as the mean of all data points in their respective clusters. This process repeats until the centroids no longer change significantly or a maximum number of iterations is reached.

### Question 3: Explain the concept of a dendrogram in hierarchical clustering?
**Answer:** A dendrogram is a tree-like diagram that visually represents the arrangement of clusters produced by hierarchical clustering. It shows the sequence of merges or splits that occur during the clustering process. The y-axis typically represents the distance or dissimilarity between clusters, while the x-axis lists the individual data points or smaller clusters. Cutting the dendrogram at a certain height reveals the clusters at that level of similarity.

### Question 4: What is the main difference between K-Means and Hierarchical Clustering?
**Answer:** The main difference is that K-Means requires the number of clusters (`k`) to be specified beforehand and produces a flat partitioning of data. Hierarchical clustering, on the other hand, does not require `k` in advance, creating a hierarchy of clusters (represented by a dendrogram) that can be cut at different levels to yield different numbers of clusters. K-Means is typically faster for large datasets, while hierarchical clustering provides more insight into the structure of the data.

### Question 5: What are the advantages of DBSCAN over K-Means?
**Answer:** DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means: it does not require specifying the number of clusters beforehand, it can find arbitrarily shaped clusters (not just spherical), and it can identify outliers as noise points. K-Means struggles with non-globular clusters and is sensitive to outliers.

### Question 6: When would you use Silhouette Score in clustering?
**Answer:** The Silhouette Score is used to evaluate the quality of clusters created by a clustering algorithm. You would use it when you need to assess how well separated the clusters are and how compactly points within a cluster are grouped. It provides a measure of how similar an object is to its own cluster compared to other clusters. A higher Silhouette Score indicates better-defined clusters.

### Question 7: What are the limitations of Hierarchical Clustering?
**Answer:** Limitations of Hierarchical Clustering include: it can be computationally expensive for large datasets (O(n^3) or O(n^2)), it has difficulty correcting mistakes made early in the merging/splitting process (greedy approach), and the choice of linkage criterion can significantly impact the results. It is also not well-suited for high-dimensional data without proper dimensionality reduction.

### Question 8: Why is feature scaling important in clustering algorithms like K-Means?
**Answer:** Feature scaling is crucial for distance-based clustering algorithms like K-Means because they rely on distance metrics (e.g., Euclidean distance) to determine the similarity between data points. If features have different scales, features with larger ranges will dominate the distance calculation, making the algorithm biased towards them. Scaling ensures all features contribute equally to the distance computation, leading to more accurate and meaningful clusters.

Question 9
How does DBSCAN identify noise points?

Answer:

DBSCAN labels points as noise (outliers) if they do not have enough neighboring points within the specified radius (eps) to satisfy the minimum number of samples (min_samples).

Question 10
Define inertia in the context of K-Means.

Answer:

Inertia is the sum of squared distances between each data point and its assigned cluster centroid. Lower inertia indicates more compact and better-defined clusters.

Question 11
What is the elbow method in K-Means clustering?

Answer:

The Elbow Method is used to determine the optimal number of clusters (K). It plots the inertia against different values of K. The point where the decrease in inertia starts to slow down (forming an "elbow") is chosen as the optimal K.

Question 12
Describe the concept of "density" in DBSCAN.

Answer:

In DBSCAN, density refers to the number of data points within a specified radius (eps) around a point. Regions with a high concentration of points form clusters, while low-density regions are treated as noise or outliers.

Question 13
Can hierarchical clustering be used on categorical data?

Answer:

Yes. Hierarchical clustering can be applied to categorical data by using appropriate similarity or distance measures, such as Hamming distance or Gower distance, instead of Euclidean distance.

Question 14
What does a negative Silhouette Score indicate?

Answer:

A negative Silhouette Score indicates that many data points are assigned to the wrong cluster and are closer to neighboring clusters than to their own cluster, suggesting poor clustering performance.

Question 15
Explain the term "linkage criteria" in hierarchical clustering.

Answer:

Linkage criteria determine how the distance between two clusters is calculated when merging them. Common linkage methods include:

Single Linkage: Minimum distance
Complete Linkage: Maximum distance
Average Linkage: Average distance
Ward Linkage: Minimizes within-cluster variance

Question 16
Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer:

K-Means assumes clusters are spherical and of similar size and density. When clusters vary significantly in size or density, it may assign points incorrectly, leading to poor clustering results.

Question 17
What are the core parameters in DBSCAN, and how do they influence clustering?

Answer:

The two main parameters are:

eps (ε): Defines the neighborhood radius around each point.
min_samples: Minimum number of neighboring points required to form a dense region.

These parameters control cluster formation and noise detection.

Question 18
How does K-Means++ improve upon standard K-Means initialization?

Answer:

K-Means++ selects initial centroids more strategically by spreading them apart. This reduces the chances of poor initialization, improves convergence speed, and often produces better clustering results.

Question 19
What is agglomerative clustering?

Answer:

Agglomerative clustering is a bottom-up hierarchical clustering method. It starts with each data point as its own cluster and repeatedly merges the closest clusters until the desired number of clusters is reached.

Question 20
What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer:

Unlike inertia, which only measures cluster compactness, the Silhouette Score evaluates both cluster cohesion and cluster separation. It provides a more comprehensive measure of clustering quality and helps compare different clustering models.

Question 21
Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

Answer:

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300,
                  centers=4,
                  random_state=42)

kmeans = KMeans(n_clusters=4,
                random_state=42)

labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1],
            c=labels,
            cmap='viridis')

plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='red',
            marker='X',
            s=200)

plt.title("K-Means Clustering")
plt.show()
````

Question 31
Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=2.5,
    random_state=42
)

model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

h = 0.05
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=model.labels_, cmap="viridis")
plt.scatter(model.cluster_centers_[:,0],
            model.cluster_centers_[:,1],
            color="red",
            marker="X",
            s=200)
plt.show()
````

Question 45
Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters.

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

X, _ = make_blobs(
    n_samples=300,
    centers=4,
    random_state=42
)

labels = AgglomerativeClustering(
    n_clusters=4,
    linkage="average"
).fit_predict(X)

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap="rainbow"
)

plt.title("Average Linkage")
plt.show()
````

Question 46
Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

````python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

wine = load_wine()

df = pd.DataFrame(
    wine.data[:, :4],
    columns=wine.feature_names[:4]
)

df["Cluster"] = KMeans(
    n_clusters=3,
    random_state=42
).fit_predict(df.iloc[:, :4])

sns.pairplot(df, hue="Cluster")

plt.show()
````

Question 47
Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, _ = make_blobs(
    n_samples=400,
    centers=4,
    cluster_std=1.8,
    random_state=42
)

labels = DBSCAN(
    eps=0.8,
    min_samples=5
).fit_predict(X)

clusters = len(set(labels)) - (1 if -1 in labels else 0)
noise = sum(labels == -1)

print("Clusters:", clusters)
print("Noise Points:", noise)
````

Question 48
Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

````python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

X = load_digits().data

X_tsne = TSNE(
    n_components=2,
    random_state=42
).fit_transform(X)

labels = AgglomerativeClustering(
    n_clusters=10
).fit_predict(X_tsne)

plt.scatter(
    X_tsne[:,0],
    X_tsne[:,1],
    c=labels,
    cmap="tab10"
)

plt.title("Agglomerative Clustering on t-SNE")
plt.show()
````
