# Question 1

**What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

**Answer:**

Anomaly Detection is the process of identifying unusual observations that differ significantly from normal data patterns.

- **Point Anomaly:** A single abnormal data point. Example: A credit card transaction of ₹5,00,000 when the customer usually spends ₹2,000.
- **Contextual Anomaly:** An observation that is abnormal only in a specific context. Example: A temperature of 25°C is normal in summer but unusual in winter.
- **Collective Anomaly:** A group of related observations that together are abnormal. Example: A sudden spike in network traffic indicating a cyberattack.

# Question 3

**What are the key components of a Time Series? Explain each with one example.**

**Answer:**

- Trend – Long-term increase/decrease. Example: Annual sales growth.
- Seasonality – Regular repeating pattern. Example: Summer ice cream sales.
- Cyclic Component – Long-term fluctuations. Example: Business cycles.
- Residual – Random variation. Example: Unexpected events.

# Question 4

**Define Stationarity in time series. How can you test and transform a non-stationary series into a stationary one?**

**Answer:**

A stationary time series has constant mean, variance, and autocorrelation over time.

**Tests:** ADF Test, KPSS Test.

**Transformations:** Differencing, log transformation, moving average, and seasonal differencing.

# Question 6

**Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.**

**Answer:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv("AirPassengers.csv", index_col="Month", parse_dates=True)
data.plot(figsize=(10,4))
plt.title("Original Time Series")
plt.show()

result = seasonal_decompose(data["Passengers"], model="multiplicative")
result.plot()
plt.show()
```

**Sample Output:** Displays the original series and its trend, seasonal, and residual components.

# Question 7

**Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

**Answer:**

```python
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,_ = make_blobs(n_samples=300, centers=1, random_state=42)
model = IsolationForest(contamination=0.05, random_state=42)
labels = model.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap="coolwarm")
plt.show()
```

**Sample Output:** Scatter plot highlighting anomalies.

# Question 8

**Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

**Answer:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = pd.read_csv("AirPassengers.csv", index_col="Month", parse_dates=True)
model = SARIMAX(data["Passengers"], order=(1,1,1), seasonal_order=(1,1,1,12))
result = model.fit()
forecast = result.forecast(12)

plt.plot(data.index, data["Passengers"], label="Actual")
plt.plot(forecast.index, forecast, label="Forecast")
plt.legend()
plt.show()
```

**Sample Output:** Forecast plot for the next 12 months.

# Question 9

**Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

**Answer:**

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,_ = make_blobs(n_samples=300, centers=1, random_state=42)
lof = LocalOutlierFactor(n_neighbors=20)
labels = lof.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap="coolwarm")
plt.show()
```

**Sample Output:** Scatter plot showing detected outliers.
